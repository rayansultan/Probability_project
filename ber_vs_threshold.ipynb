
**Analytical BER vs Decision Threshold**

This script plots the analytical bit error rate (BER) as a function
of the decision threshold V_th using the closed-form expressions
derived in the report.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

# --------- Helper functions ---------
def Phi(x):
    return 0.5 * (1 + erf(x / np.sqrt(2)))

def Q(x):
    return 1 - Phi(x)

# --------- Parameters (same as report) ---------
muZ = 1.0
sigZ = 1.0327955589886444
Delta = 1.0

# --------- Threshold range (FOCUSED range) ---------
Vth = np.linspace(0.5, 3.0, 500)

# --------- BER expression ---------
PFP = Q((Vth - muZ) / sigZ)
PFN = Phi((Vth - (muZ + Delta)) / sigZ)
BER = 0.5 * (PFP + PFN)

# --------- Plot ---------
plt.figure()
plt.plot(Vth, BER)
plt.xlabel("Decision Threshold $V_{th}$")
plt.ylabel("Bit Error Rate (BER)")
plt.title("BER as a Function of Decision Threshold")
plt.grid(True)
plt.tight_layout()
plt.show()
```
